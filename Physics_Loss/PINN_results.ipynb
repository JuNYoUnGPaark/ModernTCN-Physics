```python
import matplotlib.pyplot as plt

lambda_phys_residual = [0.1, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
f1_scores = [
    0.9719, 0.9723, 0.9731, 0.9727, 0.9744, 
    0.9772, 0.9763, 0.9763, 0.9772, 0.9771, 0.9749
]

max_f1 = max(f1_scores)
best_lambdas = [lambda_phys_residual[i] for i, score in enumerate(f1_scores) if score == max_f1]
best_scores = [max_f1] * len(best_lambdas)

plt.figure(figsize=(10, 6))
plt.plot(lambda_phys_residual, f1_scores, marker='o', linestyle='-',
         label='F1-Score', color='black')
plt.plot(best_lambdas, best_scores, 
         marker='o',          #  
         markersize=10,       #   
         color='red',         # 
         linestyle='none',    #   
         label=f'Best F1: {max_f1:.4f} (at {best_lambdas})')

plt.title('Hyperparameter Tuning: F1-Score vs. lambda_phys_residual')
plt.xlabel('lambda_phys_residual')
plt.ylabel('F1-Score (Macro)')
plt.xticks(lambda_phys_residual)
plt.ylim(0.971, 0.978)
plt.legend(loc='upper left')
plt.grid(True, linestyle='-', alpha=0.7)
plt.tight_layout()
plt.show()
```
